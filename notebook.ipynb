```python
#Load the dataset
import pandas as pd
df=pd.read_csv("zomato.csv")
```

```python
#Display the first 5 rows
print("The first 5 rows are:")
print(df.head())
```

```python
#Check missing values
df.isnull().sum()
```

```python
#Handling missing values by filling with median values
df['rate (out of 5)'] = df['rate (out of 5)'].fillna(df['rate (out of 5)'].median())
df['avg cost (two people)'] = df['avg cost (two people)'].fillna(df['avg cost (two people)'].median())
```

```python
#Removing duplicates
df = df.drop_duplicates()
```

```python
#Cleaning all text columns
df['restaurant name'] = df['restaurant name'].str.strip().str.title()
df['area'] = df['area'].str.strip().str.title()
df['local address'] = df['local address'].str.strip().str.title()
df['cuisines type'] = df['cuisines type'].str.strip().str.title()
df['restaurant type'] = df['restaurant type'].str.strip().str.title()
```

```python
#Fixing incorrect datatypes
df['rate (out of 5)'] = pd.to_numeric(df['rate (out of 5)'])
df['num of ratings'] = pd.to_numeric(df['num of ratings'])
df['avg cost (two people)'] = pd.to_numeric(df['avg cost (two people)'])
```

```python
#Renaming columns
df.columns = (df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", ""))
```

```python
#Creating new columns
df['rating_to_cost_ratio'] = df['rate_out_of_5'] / (df['avg_cost_two_people'] + 1)
df['primary_cuisine'] = df['cuisines_type'].str.split(',').str[0].str.strip()
```

```python
#Saving cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)
```
